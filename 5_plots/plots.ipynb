```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)

plt.scatter(x, y, label="sin")
plt.scatter(x, z, label="cos")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Some title")
plt.show()
```

## 6

```python
x_data = np.linspace(0, 2 * np.pi, 100)
y_data = np.sin(x_data)

fig = plt.figure()
plt.xlim(0, 2 * np.pi)
plt.ylim(-1.1, 1.1)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('sin liveplot')
line, = plt.plot([], [], lw=2)


def init():
    line.set_data([], [])
    return (line,)


def update(frame):
    x = x_data[:frame]
    y = y_data[:frame]
    line.set_data(x, y)
    return (line,)


ani = FuncAnimation(
    fig, update, frames=len(x_data) + 1, init_func=init, blit=True, interval=50
)

HTML(ani.to_jshtml())
```
